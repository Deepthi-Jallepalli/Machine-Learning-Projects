4.1 Load the first 10K samples from \train.scv" and split them into training and testing data by 0.2. Use
sklearn.svm to train a model with raw pixel features. Choose the linear kernel and set C = 5 and

 = 0:05. Report the model.

```python
import numpy as np
import pandas as p
data = p.read_csv('train.csv')
data
```

```python
X = data.iloc[0:10000,1:]
print(X)
Y = data.iloc[0:10000,0]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.svm import SVC
print('SVM Classifier with gamma = 0.05; C = 5; Kernel = linear')
classifier = SVC(C=5, gamma=0.05, kernel='linear')
classifier.fit(X_train,y_train)
```

4.2 How many support vectors does each class have? List the support vectors for each class and test the
prediction accuracy by metrics.classification report.

```python
expected = y_test
predicted = classifier.predict(X_test)
target_names = ['class 0', 'class 1', 'class 2','class 3','class 4', 'class 5', 'class 6','class 7','class 8', 'class 9']
from sklearn.metrics import classification_report
print(classification_report(expected, predicted,target_names=target_names))
```

#The above classification report shows 'support' the no.of support vectors for each class and reports accuracy for predicted values w.r.t to actuals

4.3 Repeat (4.1) and (4.2) with a RBF kernel and the same C and gamma

```python
print('SVM Classifier with gamma = 0.05; C = 5; Kernel = linear')
classifier = SVC(C=5, gamma=0.05, kernel='rbf')
classifier.fit(X_train,y_train)
```

```python
expected = y_test
predicted = classifier.predict(X_test)
target_names = ['class 0', 'class 1', 'class 2','class 3','class 4', 'class 5', 'class 6','class 7','class 8', 'class 9']
from sklearn.metrics import classification_report
print(classification_report(expected, predicted,target_names=target_names))
```
